```python
data = pd.read_excel(r"D:\KHO\Cuoi_ki\data-K-Mean.xlsx")
data
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt
data.plot(kind='scatter',x='Length',y='Width')
plt.show()
```

```python
import sklearn.cluster as cluster
kmeans = cluster.KMeans(n_clusters=3)
kmeans = kmeans.fit(data[['Length','Width']])
```

```python
centroids=kmeans.cluster_centers_
centroids
```

```python

data.insert(column='Cluster',value=kmeans.labels_,loc=3)
data
```

tạo một thuộc tính cluster(final),value : các giá trị của cột để thêm, loc=3: vị trí cột số 3

```python
data['Cluster'].value_counts()
```

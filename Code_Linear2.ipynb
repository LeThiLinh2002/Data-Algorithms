```python
Radio=pd.DataFrame (data['Radio'])
Sales=pd.DataFrame (data['Sales'])
```

```python
Radio
```

```python
from sklearn import linear_model
```

```python
lm=linear_model.LinearRegression()
model=lm.fit(Radio,Sales)
```

```python
model.coef_
```

coef: tính hệ số góc (=a)

```python
model.intercept_
```

intercept: hàm tính hệ số b: hệ số điều chỉnh

```python
model.score(Radio,Sales)
```

```python
Radio_new=[[40]]
Sales_predict=model.predict(Radio_new)
Sales_predict
```

```python
data.plot(kind='scatter',x='Radio',y='Sales')

plt.plot(Radio,model.predict(Radio),color='red',linewidth=2)
plt.scatter(Radio_new, Sales_predict, color='yellow')
plt.show()
```
